# タイムゾーンの変換
`日時処理`, `タイムゾーン`

`astimezone`メソッドを使う。  
引数は `pytz.timezone` オブジェクトである。  


````python
from dateutil.parser import parse
from pytz import timezone
from IPython.display import display
timestamp_desc = '2019-05-13T12:00:00+09:00'
print("[受け取った日付のdatetime型]\n{}\n".format(parse(timestamp_desc)))

# UTC に変換してみる
timestamp_utc = parse(timestamp_desc).astimezone(timezone('UTC'))
print("[UTCに変換した後のdatetime型]\n{}\n".format(timestamp_utc))
````

## データフレームインデックスのタイムゾーン変換
`tz_convert`と`tz_localize`が用意されている。  
`pd.date_range`で生成したタイムスタンプ列は、デフォルトでは`naive`つまりタイムゾーン属性が付与されていない。  
そのため、`tz_convert`を実行するとエラーが発生する。  
```python
TypeError: Cannot convert tz-naive timestamps, use tz_localize to localize
```

`tz_localize`をすれば、タイムゾーンが付与されたタイムスタンプ列に変換される。  
この状態になって以降であれば、`tz_convert`で任意のタイムゾーンに変換することが出来るようになる。　　

* [参考:pandasの時系列データのタイムゾーンを処理（tz_convert, tz_localize） | note.nkmk.me](https://note.nkmk.me/python-pandas-tz-convert-tz-localize/)

````python
import pandas as pd 
df_tmp = pd.DataFrame({'val':range(0,20)}, index=pd.date_range('2019-05-01 00:00',periods=20,freq='H'))
````

````python
# naive な状態のタイムスタンプ列をローカルタイムスタンプに変換する
df_tmp.index.tz_localize('Asia/Tokyo')
````

````python
# tz_localize でタイムゾーンを付与した後であれば、tz_convertで任意のタイムゾーンに変換することが出来る。
df_tmp.index.tz_localize('Asia/Tokyo').tz_convert('UTC')
````

# マルチレベルカラムの操作
## マルチレベルカラムを定義する
今あるデータフレームに、階層を一つ追加したい際に便利。  
* [pandas.MultiIndex.from_arrays](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.MultiIndex.from_arrays.html)

引数は、各階層ごとに見た際の各カラムのラベルを格納した二次元の配列になる。   
以下は、N個のカラムが、M階層構造をしている場合の配列の構造である。

|カラム１|カラム２|...|カラムN|
|---|---|---|---|
|$Column_{1}^{level1}$|$Column_{2}^{level1}$|...|$Column_{N}^{level1}$|
|$Column_{1}^{level2}$|$Column_{2}^{level2}$|...|$Column_{N}^{level2}$|
|...|...|...|...|
|$Column_{1}^{levelM}$|$Column_{2}^{levelM}$|...|$Column_{N}^{levelM}$|

$a_1$

````python
df_tmp = pd.DataFrame({'a': range(100,120), 'b':range(120,140), 'c':range(140,160)})
````

変更前のデータフレームは、単階層カラムである。

````python
df_tmp.head()
````

3階層カラムは、以下のように定義することが出来る。  
`pd.MultiIndex.from_arrays`を使っている。  
その第一引数には、階層カラムの定義を二次元リストで指定している。  
* 各列が、カラムに対応している。
* 各行は、階層に対応している。

````python
#                   対応する元カラム        a          b           c
multi_cols = pd.MultiIndex.from_arrays([['USD/JPY', 'EUR/JPY', 'GBP/JPY'],
                                        ['Price',   'Price',   'Price'],
                                        ['Bid',     'Ask',     'Bid']],
                                      names=('Commodity', 'Value Type', 'Bid/Ask'))
df_tmp.columns = multi_cols
````

# 移動相関係数の計算
`pd.Series.rolling.corr`を使う。  
窓を動かしながら、窓に含まれる数値から相関係数を求める。  
corr() の引数には、相関係数を求める際の相手方変数を指定する。  
指定しないと、**自分自身**との相関係数を求めてしまう。  

* [参考:pandas.core.window.Rolling.corr](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.window.Rolling.corr.html)

````python
df_tmp = pd.DataFrame({'pos': range(0,100),
                      'value_x': np.random.rand(100),
                      'value_y': np.random.rand(100)})
````

````python
# 窓の幅は10(10レコード)、value_xとvalue_yの相関係数を求める
df_tmp['corr'] = df_tmp['value_y'].rolling(window=10).corr(df_tmp['value_x'])
````

````python
df_tmp
````

````python
df_tmp['corr_self'] = df_tmp['value_y'].rolling(window=10).corr()
````

````python
df_tmp
````

## プログレスバーの表示方法
`print` 関数の引数に `end=''` を指定した上で、出力末尾に `\r` を付与すれば良い。  
表示位置が行頭に戻り、なおかつ改行が出力されないので、プログレスバーを上書きし続けることが可能となる。  

````python
import sys, time
 
def prog_bar(length=100):
    for i in range(length+1):
        num_spc=100 - i
        print('[' + '#'*i + ' ' * num_spc + ']' + '{}/{}'.format(i,100) + '\r', end='')
        time.sleep(0.01)
    print("")
 
for i in range(4):
    prog_bar()
````
